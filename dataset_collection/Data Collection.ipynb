```python
frequency = 'weekly'
basedir = '../data_countries/'


for country, code in zip(countries, country_codes):
    path = basedir+country+"/"
    print(country)
    try:
        os.mkdir(path)
    except:
        print("directory exists")
    for date in dates:
        url = 'https://spotifycharts.com/viral/{}/weekly/{}/download'.format(code,date)
        filename = '{}-{}'.format(country,date)
        re = requests.get(url)
        if str.encode("<!doctype html>") in re.content:
            print("Error, skipping file:")
            print(url)
            print()
            continue
        else:
            with open(path+filename, 'wb') as f:
                f.write(re.content)
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=credentials.c_id, client_secret=credentials.c_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
base = '../data_countries'
master = pd.DataFrame()

# Directories
for directory in os.listdir(base):
    path = base + '/' + directory
    print(path)
    
    #Files
    alpha = pd.DataFrame()
    for current_file in os.listdir(path):
        file_path = path + '/' + current_file
        beta = pd.read_csv(file_path)
        alpha = alpha.append(beta)
        
    master = master.append(alpha)
master = master.drop(columns=['Position'])
master = master.drop_duplicates(subset="URL")
master = master.reset_index(drop=True)
```

```python
batches_number = round(len(master)/100)
features = pd.DataFrame()
removed_songs = []

for batch_number in range(1, batches_number + 1):
    min_val = (batch_number - 1) * 100
    max_val = min(batch_number * 100, len(master))
    batch = master[min_val:max_val]["URL"].tolist()
    
    # Get features
    feat = sp.audio_features(batch)
    
    # Find url of none songs
    none_songs = [i for i in range(len(feat)) if feat[i] == None]
    for none_song in none_songs:
        removed_songs.append(master.iloc[min_val + none_song]['URL'])
    
    # remove none songs from features
    feat = [i for i in feat if i]
    features = features.append(feat)


    if batch_number % 100 == 0:
        print(batch_number)
features = features.reset_index(drop=True)
```

```python
for removed_song in removed_songs:
    master = master[master.URL != removed_song]
master = master.reset_index(drop=True)
master = pd.concat([master, features], axis=1)
master = master.drop(columns=['id', 'type', 'uri', 'track_href', 'analysis_url'])
master.to_csv("song_masterlist.csv", index=False)
```

```python
# Directories
for directory in os.listdir(base):
    path = base + '/' + directory
    print(path)
    
    #Files
    alpha = pd.DataFrame()
    for current_file in os.listdir(path):
        file_path = path + '/' + current_file
        beta = pd.read_csv(file_path)
        before_len = len(beta)
        for removed_song in removed_songs:
            beta = beta[beta.URL != removed_song]
        after_len = len(beta)
        if before_len != after_len:
            print("Removed from: " + file_path)
        beta.to_csv(file_path, index=False)
```
